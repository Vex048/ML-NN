```python
df=df.drop(['country'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
columns_to_encode=['statezip','city','street']
for col in columns_to_encode:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
Y=df['price']
df=df.drop(['price'],axis=1)
categorical_columns=['waterfront','view','condition']
X=pd.get_dummies(df,columns=categorical_columns)
X=X.replace({True:1,False:0})
X
```

```python
X=X.drop(['date'],axis=1)
```

```python
cols_to_scale=['bedrooms','bathrooms','sqft_living','sqft_lot','floors','sqft_above','sqft_basement','yr_built','yr_renovated','street','city','statezip']
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=X
X_scaled[cols_to_scale]=scaler.fit_transform(X_scaled[cols_to_scale])
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
```

```python
print("Score of the model is: ",model.score(X_test,y_test))
from sklearn.metrics import mean_squared_error
score= mean_squared_error(y_test,y_predict)
print("MSE of the model is: ",score)
```

```python
x=X_test.shape[0]
x=np.arange(0,x,1,dtype=int)
plt.plot(x,y_test,color='b',label='test')
plt.plot(x,y_predict,color='r', label='pred')
plt.show()
```

```python
cols=list(X_scaled.columns)
X_to=X[cols]
X_to
X_to=X_to.to_numpy()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(2)
X_poly=X_scaled
X_poly=poly.fit_transform(X_poly[cols_to_scale])
X_poly=np.concatenate([X_poly,X_to],axis=1)
X_poly.shape
```

```python
X_train_poly,X_test_poly,y_train,y_test = train_test_split(X_poly,Y,test_size=0.2,random_state=42)
```

```python
model_poly2=LinearRegression()
model_poly2.fit(X_train_poly,y_train)
y_predict2=model_poly2.predict(X_test_poly)
```

```python
score_poly2= mean_squared_error(y_test,y_predict2)
print("MSE of the Polynominal model with 2 degree is: ",score_poly2)
print("Score of the model with polynominal degree is: ",model_poly2.score(X_test_poly,y_test))
```
